```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv('/content/synthetic_study_data.csv')
```

```python
data.head(50)
```

```python
data.info()
```

```python
# to see columns heading
print("Columns available:", data.columns.tolist())
```

```python
# basic feature
feature=['Hours_Studied']
```

```python
# checking if there exist
for col in feature:
    if col not in data.columns:
        raise ValueError(f"Column '{col}' not found in dataset!")
```

```python
x=data['Hours_Studied'].values
y=data['Marks'].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# scaling (standardizing)
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train.reshape(-1, 1))# we reshape into (-1,1) because scaler transformation takes place in 2d form
x_test=scaler.transform(x_test.reshape(-1, 1))# (-1,1) create [[1,2,3]]into [[1][2][3]]
```

```python
model = models.Sequential([
    layers.Dense(16, activation='relu', input_shape=(x_train.shape[1],)),

    layers.Dense(1)  # Regression output
])
```

```python
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
#under model.fit:
"""for epoch in range(epochs):
    predictions = forward_pass(X)
    loss = compute_loss(y, predictions)
    gradients = backpropagation(loss)
    update_weights(weights, gradients, learning_rate)"""

history = model.fit(
    x_train, y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=8,
    verbose=1
)
```

```python
preds = model.predict(x_test).flatten()# single output neuron (Dense(1)) so we faltten it
mse = mean_squared_error(y_test, preds)
print(f"Test MSE: {mse:.2f}")
```

```python
#  print actual vs predicted
for actual, pred in zip(y_test[:5], preds[:5]):
    print(f"Actual: {actual:.2f}, Predicted: {pred:.2f}")
```

```python
import numpy as np

# Step 1: Get user input
hours_studied = float(input("Enter hours studied: "))

# Step 2: Prepare it as a 2D array (because model expects shape (n_samples, n_features))
new_data = np.array([[hours_studied]])

# Step 3: Apply same scaling as training
new_data_scaled = scaler.transform(new_data)

# Step 4: Predict marks
predicted_marks = model.predict(new_data_scaled)

# Step 5: Print result
print(f"Predicted Marks: {predicted_marks[0][0]:.2f}")

```
